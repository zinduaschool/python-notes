```python
# we want to make some changes to our googl sheet
gc = gspread.service_account(filename='google_sheet_creds.json') # We load the API credentials created in Google
sp =gc.open('chris_spreadsheet').sheet1 # We open Sheet 1
sp.append_row([x for x in range(1,10)]) # Adding Data row-wise, the more you run this piece of code, the more the number of rows are added to your data.

# we can also use the update method
# create a 10 by 10 list matrix
matrix = [[random.randint(1, 100) for _ in range(10)] for _ in range(10)]
sp.update(matrix,'A5') # We will use cell A5 as the first point.
```

Now, we have explored obtaining API keys, using gspread to update our Spreadsheets, we now want to  get information from open countries rest api (CountryLayer to be specific. This will require that we create API Keys)

```python
base_url = 'https://api.countrylayer.com/v2/'
endpoint = base_url+'all'
params={'access_key':'33e5d4c716bef4d12d0127a0ccd39e97'}
response = requests.get(endpoint,params=params)
```

```python
response.json()[14]
```

```python
countries = response.json()
data = [['Name','Alpha2Code','Capital_City','Continent','CallingCodes','altSpellings']]

for country in countries:
    name = country['name']
    alpha2Code = country['alpha2Code']
    capital = country['capital']
    region = country['region']
    callingCodes = ", ".join(country['callingCodes'])
    altSpellings = "; ".join(country['altSpellings'])
    data.append([name,alpha2Code,capital,region,callingCodes,altSpellings])
data[117]
```

```python
# put it in the sheet
sp.update(data,'A1')
```
